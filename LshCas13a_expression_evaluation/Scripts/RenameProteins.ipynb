```python
E_coli_proteins = "Annotations/E_coli_proteins.fasta"
```

```python
from Bio import SeqIO
from Bio.SeqRecord import SeqRecord
```

```python
ProteinsRecords = [i for i in SeqIO.parse(E_coli_proteins, format="fasta")]
```

```python
ProteinsRecords
```

```python
import re
```

```python
EditedProteinRecords = []

for prot in ProteinsRecords:
    NewProtenRecord = SeqRecord(id = re.search("\[gene=(.*?)\]", prot.description).group(1), 
                                seq = prot.seq, 
                                description="")
    EditedProteinRecords.append(NewProtenRecord)
```

```python
SeqIO.write(EditedProteinRecords, "Annotations/E_coli_proteins_edited.fasta", format="fasta")
```
