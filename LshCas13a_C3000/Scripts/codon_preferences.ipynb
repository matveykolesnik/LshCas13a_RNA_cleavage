```python
TCS_across_features_file = "Results/Tables/TCS_detection_tables/LRTest_table_and_genome_features.tsv"
Reference_sequences_file = "Reference_sequences/Targeting_samples_refseqs.fasta"
```

```python
import pandas as pd
```

```python
TCS_across_features_DF = pd.read_csv(TCS_across_features_file, sep="\t")
TCS_across_features_DF
```

```python
TCS_across_ORFs = TCS_across_features_DF.query("MatchedFeatureType == 'CDS' and logFC > 0").sort_values("PValue.adj")
TCS_across_ORFs
```

```python
for i in zip(TCS_across_ORFs["Strand"], TCS_across_ORFs["MatchedFeatureStrand"]):
    if i[0] != i[1]:
        print("Some TCS strand does not fit matched ORF strand")
```

```python
def calc_distance_from_init_codon(row):
    if row.Strand == '+':
        return(row.Pos - row.MatchedFeatureStart + 1)
    else:
        return(row.MatchedFeatureEnd - row.Pos + 1)
```

```python
TCS_across_ORFs["DistFromInitCodon"] = TCS_across_ORFs.apply(calc_distance_from_init_codon, axis=1)
TCS_across_ORFs
```

```python
def get_codon_information(row):
    q = row.DistFromInitCodon // 3
    r = row.DistFromInitCodon % 3
    if r > 0:
        codon_position = q+1
    else:
        codon_position = q
    TCS_position_within_codon = 3 - r
    return((codon_position, TCS_position_within_codon))
```

```python
TCS_across_ORFs[["codon_number", "pos_within_codon"]] = TCS_across_ORFs.apply(get_codon_information, axis=1, result_type="expand")
```

```python
from Bio import SeqIO
```

```python
SeqRecords = [i for i in SeqIO.parse(Reference_sequences_file, "fasta")]
SeqRecords
```

```python
SeqDict = dict((i.id, i.seq) for i in SeqRecords)
SeqDict
```

```python
SeqDict["NC_000913.3"][10:20]
```

```python
def extract_codon(row, seqdict):
    r_side = 3 - row.pos_within_codon
    l_side = 3 - r_side - 1
    
    if row.Strand == "+":
        codon_sequence = str(seqdict[row.SeqID][row.Pos-l_side-1 : row.Pos+r_side]).upper().replace("T", "U")
    else:
        codon_sequence = str(seqdict[row.SeqID][row.Pos-r_side-1 : row.Pos+l_side].reverse_complement()).upper().replace("T", "U")
        
    return(codon_sequence)
```

```python
TCS_across_ORFs["codon_seq"] = TCS_across_ORFs.apply(extract_codon, args=[SeqDict], axis=1)
TCS_across_ORFs
```

```python
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
```

```python
N=1000
TCS_across_ORFs_subset = TCS_across_ORFs.head(N)
TCS_across_ORFs_subset
```

```python
codon_cleavage_frequencies = Counter(TCS_across_ORFs_subset.codon_seq)
codon_cleavage_frequencies
```

```python
codon_cleavage_frequencies_sorted = sorted(codon_cleavage_frequencies.items(), key=lambda x: x[1], reverse=True)
codon_cleavage_frequencies_sorted
```

```python
codons, counts = zip(*codon_cleavage_frequencies_sorted)
print(codons)
print(counts)
```

```python
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['mathtext.fontset'] = 'custom'
plt.rcParams['mathtext.it'] = 'Arial:italic'
plt.rcParams['mathtext.bf'] = 'Arial:italic:bold'
```

```python
fig, ax = plt.subplots(figsize=(40,10), facecolor="white")
ax.bar(x=codons, 
       height=counts, 
       color="red", 
       edgecolor="black")

ax.tick_params(labelsize=20)
ax.tick_params(axis="x", rotation=45)
ax.set_ylabel("Number of mRNA cleavages in the codon", size=30)
ax.set_xlabel("Codon sequence", size=30)

fig.savefig("Results/Pictures/codon_cleavage_preferences.png", dpi=300, bbox_inches="tight", transparent=False)
```

```python
codon_cleavage_frequencies_positions = Counter(zip(TCS_across_ORFs_subset.codon_seq, TCS_across_ORFs_subset.pos_within_codon))
codon_cleavage_frequencies_positions_sorted = sorted(codon_cleavage_frequencies_positions.items(), key=lambda x: x[1], reverse=True)
codon_cleavage_frequencies_positions_sorted
```

```python
codon_pos_counts_list= [(*cp, count) for cp, count in codon_cleavage_frequencies_positions_sorted]
codon_pos_counts_list
```

```python
source_df = pd.DataFrame({"codon" : codons, 
              "1" : np.zeros(len(codons)), 
              "2" : np.zeros(len(codons)), 
              "3" : np.zeros(len(codons))})

for i in codon_pos_counts_list:
    source_df.loc[source_df.codon == i[0], str(i[1])] = i[2]
    
source_df
```

```python
fig, ax = plt.subplots(figsize=(40,10), facecolor="white")

#cleavage between 2 and 3
ax.bar(x=codons, 
       height=source_df["3"], 
       color="red", 
       edgecolor="black", 
       label="2/3")

#cleavage between 1 and 2
ax.bar(x=codons, 
       height=source_df["2"],
       bottom=source_df["3"],
       color="green", 
       edgecolor="black", 
       label="1/2")

#cleavage between -1 and 1
ax.bar(x=codons, 
       height=source_df["1"],
       bottom=source_df["2"] + source_df["3"],
       color="yellow", 
       edgecolor="black", 
       label="-1/1")

handles, labels = ax.get_legend_handles_labels()
ax.legend(reversed(handles),
          reversed(labels),
          fontsize=20, 
          prop = {"size" : 30}, 
          title = "Cleavage positions within codons", 
          title_fontsize=30)

ax.tick_params(labelsize=20)
ax.tick_params(axis="x", rotation=45)
ax.set_ylabel("Number of mRNA cleavages in the codon", size=30)
ax.set_xlabel("Codon sequence", size=30)

fig.savefig("Results/Pictures/codon_cleavage_preferences_with_pos.png", dpi=300, bbox_inches="tight", transparent=False)
```
